```python
#Now we create a train test split for the data
from sklearn.model_selection import train_test_split

y = housing_df['Price']
X = housing_df.drop(['Price', 'Address'], axis = 1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
###################################################### BUILDING MODELS #######################################################
```

```python
#Out of the three metrics MAE, MSE, and RMSE
# MAE is the easiest to understand
# MSE is the more pupular than MAE beacause MSE "punishes" larger errors which is useful in real-world scenerio
# RMSE is the most popular out of the three because RMSE is interpretable in the "y" units

#Another metrics we use is coefficient of determination or R2 score, which is the percentage of variation in y explained by all
# the x variables together. Usually, an R2 of .7 is considered good
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)

predictions_linear = lm.predict(X_test)
```

```python
#Accuracy Metrics
from sklearn import metrics
from sklearn.metrics import r2_score

# printing metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions_linear))
print('MSE:', metrics.mean_squared_error(y_test, predictions_linear))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions_linear)))
print('R2 Score: ', r2_score(y_test, predictions_linear))

#With an R2 score 0f 0.9 seems like the linear model is a good model.
```

```python
#Ridge Regression Model
# Ridge Regression is a technique for analyzing multiple regression data that suffers from multicollinearity. When multicollinarity
# occurs, least squares estimates are unbiased, but their variances are large so they may be far from the true value. By adding
# a degree of bias to the regression estimates, ridge regression redcues the standard errors. It is hoped that the net effects will be
# to give estimates that are more reliable.

from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 0.05, normalize = True)
ridge.fit(X_train, y_train)

predictions_ridge = ridge.predict(X_test)

# printing metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions_ridge))
print('MSE:', metrics.mean_squared_error(y_test, predictions_ridge))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions_ridge)))
print('R2 Score: ', r2_score(y_test, predictions_ridge))
```

```python
#Lasso Regression Model
# The "LASSO" stands for Least Absolute Shrinkage and Selection Operator. This model uses shrinkage. Shrinkage is where data 
# values are shrunk towards a central point as the mean. The lasso procedure encourages simple, sparse models (i.e. models with
# fewer parameters). This particular type of regression is well-suited for models showing high levels of multicollinarirty or when
# you want to automate certain parts of model selection, like variable/parameter elimination.

#The key difference is Lasso shrinks the less important feature's coefficient to zero, thus removing some feature altogetherr.
# So, this works well for feature selection incase we have ahuge number of features.

from sklearn.linear_model import Lasso

lasso = Lasso(alpha = 0.5, normalize = True)
lasso.fit(X_train, y_train)

predictions_lasso = lasso.predict(X_test)

# printing metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions_lasso))
print('MSE:', metrics.mean_squared_error(y_test, predictions_lasso))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions_lasso)))
print('R2 Score: ', r2_score(y_test, predictions_lasso))

#Based on the results from all three models (linear, ridge, and lasso), all three models performed well with an accuracy rate >90%
# Out of the three, the linear has a slight edge overall with slightly higher R2 score and slighty less MAE. Ridge seems to performed
# the worst out of the three, but not by much.
```

```python
######################################## IMPROVING LINEAR, RIDGE, AND LASSO MODELS ############################################
```

```python
#Applying 10-fold cross validation to the previous linear model to see if we can get a better result

from sklearn.model_selection import cross_val_score
lm2 = LinearRegression()
scores_lm = cross_val_score(lm2, X_train, y_train, cv = 10)
print('Cross-validation scores:{}'.format(scores_lm))

#Compute average score
print('\nAverage Cross-validation score: {:.4f}'.format(scores_lm.mean()))
```

```python
#Same with Ridge
ridge2 = Ridge(alpha = 0.05, normalize = True)
scores_ridge = cross_val_score(ridge2, X_train, y_train, cv = 10)
print('Cross-validation scores:{}'.format(scores_ridge))

#Compute average score
print('\nAverage Cross-validation score: {:.4f}'.format(scores_ridge.mean()))
```

```python
#And with Lasso
lasso2 = Lasso(alpha = 0.05, normalize = True)
scores_lasso = cross_val_score(lasso2, X_train, y_train, cv = 10)
print('Cross-validation scores:{}'.format(scores_lasso))

#Compute average score
print('\nAverage Cross-validation score: {:.4f}'.format(scores_lasso.mean()))
```
