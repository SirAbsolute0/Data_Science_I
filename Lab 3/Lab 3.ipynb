```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
#Setting figure size preference and visual mode
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 5 #inches
sns.set_style('darkgrid')
```

```python
#Example of a simple graph
x = np.linspace(0, 20)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
#add title and labels
# plt.tilte()
# plt.x/ylabel()

plt.plot(x, y)
plt.title('My Graph Title')
plt.xlabel('My x-axis label')
plt.ylabel('My y-axis label')
plt.show()
```

```python
#Plotting two lines in the same graph

plt.plot(x, np.sin(x), label = 'y=sin(x)')
plt.plot(x, np.cos(x), label = 'y=cos(x)')
plt.title("Sin and Cos Functions")
plt.xlabel("x-axis")
plt.ylabel('y-axis')

plt.legend()
plt.show()
```

```python
# If we want to seperate the plot of sin(x) and cos(x) into seperate plots but side by side, we use subplot(nrows, ncols)
# Imagine seperating one big white canvas (like a 1x1) into two graphs onto 1 white canvas (like a 1x2)

fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, np.sin(x))
axes[0].set_title('y = sin(x)')

axes[1].plot(x, np.cos(x))
axes[1].set_title('y = cost(x)')

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (14, 8))

axes[0][0].plot(x, np.sin(x))
axes[0][0].set_title('y = sin(x)')

axes[1][1].plot(x, np.cos(x))
axes[1][1].set_title('y = cost(x)')

print(axes.shape)
print(type(fig))
plt.show()
```

```python
#We can also customize the x and y axis range
# set_ylim([y_start, y_end])
# set_xlim([x_start, x_end])
# axis('tight') - this is for automatically getting "tightly fitted" axes ranges

x = np.linspace(0, 5, 11) #From 0 to 5 with total 11 values in the range
fig, axes = plt.subplots(nrows = 1, ncols = 3)

axes[0].plot(x, x**2, x**4)
axes[0].set_title('default axis range')

axes[1].plot(x, x**2, x**4)
axes[1].axis('tight')
axes[1].set_title('Using axis("tight")')

axes[2].plot(x, x**2, x**4)
axes[2].set_ylim([0, 40])
axes[2].set_xlim([2, 3])
axes[2].set_title('Custom axes range')

plt.show()
```

```python
pokemon_df = pd.read_csv('Pokemon.csv')
pokemon_df.head()
```

```python
#Using scatter plot
# plt.scatter(x, y)

plt.scatter(x = pokemon_df['HP'], y = pokemon_df['Attack'])
plt.xlabel('HP')
plt.ylabel('Attack')
plt.title('HP vs Attack')
plt.show()
```

```python
# If we have seaborn, we can even expand the above graph by plotting each point with a specific category.
sns.scatterplot(x = pokemon_df['HP'], y = pokemon_df['Attack'], hue = 'Generation', data = pokemon_df)
plt.show()
```

```python
# Using histogram
# plt.hist(x)
plt.hist(x = pokemon_df['Attack'])
plt.show()
```

```python
#Note since .value_counts() is like creating a dictionary where the key is on the left and the value is on the right,
# we can get the keys with .index and get all the values with .values
pokemon_df['Type 1'].value_counts()
```

```python
#Calling .index of of .value_counts() will return just the index names
#Calling .values off of .value_counts() will return just the corresponding values
#We can use both information to creat a bar plot (plt.bar(x, y)) with x as our index and y as our values

type_1_names = pokemon_df['Type 1'].value_counts().index
type_1_values = pokemon_df['Type 1'].value_counts().values

plt.bar(type_1_names, type_1_values)
plt.title('Type 1 Pokemon value counts')
plt.show()
```

```python
#Using pie chart
# From the bar graph above, we can take the top 5 types and represent in a pie chart
# plt.pie(sizes, colors, Shadow = True (add a slight shadow for visual), startangle = 90 (the start angle of the pie))
# plt.pie(sizes, colors) return two things: patchs - a sequence of matplotlib.patches.wedge instances and texts - a list of the label text instances
# It is important to save these two variables to use with plt.legend(patches, names)

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'brown']
patches, texts = plt.pie(type_1_values[:5], colors = colors, shadow = True, startangle=90) #colors will be randomized if not specified
plt.legend(patches, type_1_names[:5])

plt.axis('equal') #To center the pie chart
plt.tight_layout()
plt.show()
```

```python
#By collecting the first 4 pokemon names in our DataFrame along with their HP, Attack, and Defense
# We can compare these 4 pokemon using a stack plot
# plt.stackplot(x, y1, y2, y3) - x is the 4 pokemon, y1 is HP, y2 is Attack, y3 is Defense

x = pokemon_df['Name'][:4]
y1 = pokemon_df['HP'][:4]
y2 = pokemon_df['Attack'][:4]
y3 = pokemon_df['Defense'][:4]

labels = ["HP", "Attack", "Defense"]
plt.stackplot(x, y1, y2, y3)
plt.legend(loc = 'upper left', labels = labels)

plt.xticks(rotation = 90) #rotate the label to appear vertical rather than horizontal by default
plt.show()
```

```python
#Representing the stackplot in terms of subplot

fig, axes = plt.subplots(nrows = 3, ncols = 1, sharey = True) #Using sharey makes that we don't have to create sperate .set_tilte, set_ylabel, etc
axes[0].plot(x, y1, 'ko-')
axes[0].set(title = '3 subplots', ylabel = 'HP')

axes[1].plot(x, y2, 'r.-')
axes[1].set(ylabel = 'Attack')

axes[2].plot(x, y3, ':')
axes[2].set(xlabel = 'Pokemon', ylabel = 'Defense')

plt.show()
```

```python
#We can also make line plots of panda columns which we have seen before
plt.plot(pokemon_df['Attack'], label = ['Attack'], color = 'red')
plt.plot(pokemon_df['Defense'], label = ['Defense'], color = 'green')
plt.plot(pokemon_df['HP'], label = ['HP'], color = 'blue')
plt.legend()
plt.show()
```

```python
#Heatmap: if we call .corr() on a DataFrame, we will get back a correlation matrix
# P.S: Make sure you don't have categorical variables in your DataFrame or you will get an error (the values must be number)
cor_value = pokemon_df[['HP', 'Attack', 'Defense']].corr()
cor_value
```

```python
#We can then pass this information into sns.heatmap() to visualize it
# Can also use matplotlib for this but seaborn is easier with only 1 line of code
sns.heatmap(cor_value)
plt.show()
```
