```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
#Setting figure size preference and visual mode
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 5 #inches
sns.set_style('darkgrid')
```

```python
#Example of a simple graph
x = np.linspace(0, 20)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
#add title and labels
# plt.tilte()
# plt.x/ylabel()

plt.plot(x, y)
plt.title('My Graph Title')
plt.xlabel('My x-axis label')
plt.ylabel('My y-axis label')
plt.show()
```

```python
#Plotting two lines in the same graph

plt.plot(x, np.sin(x), label = 'y=sin(x)')
plt.plot(x, np.cos(x), label = 'y=cos(x)')
plt.title("Sin and Cos Functions")
plt.xlabel("x-axis")
plt.ylabel('y-axis')

plt.legend()
plt.show()
```
